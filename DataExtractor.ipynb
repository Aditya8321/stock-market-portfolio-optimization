```python
import yfinance as yf
import pandas as pd
import ta

stock_data=pd.read_csv("/content/SUZLON.csv")

# Calculate the Exponential Moving Averages (EMA)
stock_data['EMA_50'] = stock_data['Close'].ewm(span=50, adjust=False).mean()
stock_data['EMA_200'] = stock_data['Close'].ewm(span=200, adjust=False).mean()

stock_data['EMA_diff'] = stock_data['EMA_50'] - stock_data['EMA_200']

rsi_window = 14
stock_data['RSI'] = ta.momentum.RSIIndicator(stock_data['Close'], window=rsi_window).rsi() / 100
stock_data = stock_data.dropna()
stock_data.to_csv("SUZ_features.csv")
```
