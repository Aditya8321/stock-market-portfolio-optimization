```python
import pandas as pd

# Load the CSV files
file1 = pd.read_csv('/content/PNB_features.csv')  # Replace with your file names
file2 = pd.read_csv('/content/PNB_features.csv')
file3 = pd.read_csv('/content/TATAMOTORS_features.csv')

# Convert the date columns to datetime format if necessary
# Merge the dataframes on the 'date' column
merged_data = pd.merge(file1, file2, on='Date', how='inner')  # Use 'outer', 'inner', 'left', or 'right' as needed
merged_data = pd.merge(merged_data, file3, on='Date', how='inner')

# Save the merged dataframe to a new CSV file
merged_data.to_csv('merged_output.csv', index=False)

print("Files merged successfully!")
```
